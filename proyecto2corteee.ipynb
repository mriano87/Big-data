```python
import requests
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import wbgapi as wb
import plotly.express as px
import numpy as np
```

```python
PIB_per_capita = wb.data.DataFrame('NY.GDP.PCAP.CD',
                                  economy={'COL',"BRA", "CRI" ,"ECU", "ARG" , "USA", "MEX", "PER", "PAN","CAN", "URY", "PRY", "CHL", "BOL"},
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index("Country").unstack().reset_index()
```

```python
PIB_per_capita
```

```python
Esperanza = wb.data.DataFrame('SP.DYN.LE00.IN',
                                  economy={'COL',"BRA", "CRI" ,"ECU", "ARG" , "USA", "MEX", "PER", "PAN","CAN", "URY", "PRY", "CHL", "BOL"},
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index("Country").unstack().reset_index()
```

```python
Esperanza
```

```python
Inflacion = wb.data.DataFrame('FP.CPI.TOTL.ZG',
                                  economy={'COL',"BRA", "CRI" ,"ECU", "ARG" , "USA", "MEX", "PER", "PAN","CAN", "URY", "PRY", "CHL", "BOL"},
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index("Country").unstack().reset_index()
Inflacion
```

```python
Desempleo = wb.data.DataFrame('SL.UEM.TOTL.ZS',
                                  economy={'COL',"BRA", "CRI" ,"ECU", "ARG" , "USA", "MEX", "PER", "PAN","CAN", "URY", "PRY", "CHL", "BOL"},
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index("Country").unstack().reset_index()
Desempleo
```

```python
Mortalidad_infantil = wb.data.DataFrame('SH.DYN.MORT',
                                  economy={'COL',"BRA", "CRI" ,"ECU", "ARG" , "USA", "MEX", "PER", "PAN","CAN", "URY", "PRY", "CHL", "BOL"},
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index("Country").unstack().reset_index()
Mortalidad_infantil
```

```python
Tabla_final=(Mortalidad_infantil.merge(Inflacion,how="outer",on=["level_0","Country"])
                    .merge(Esperanza,how="outer",on=["level_0","Country"])
                    .merge(PIB_per_capita,how="outer",on=["level_0","Country"])
                    .merge(Desempleo,how="outer",on=["level_0","Country"]))
```

```python
Tabla_final.columns
```

```python
(Mortalidad_infantil.merge(Inflacion,how="outer",on=["level_0","Country"]))
```

```python
Tabla_final.columns= ["Año","Pais","Mortalidad_infantil","Inflacion","Esperanza","PIB_per_capita","Desempleo"]
Tabla_final
```
